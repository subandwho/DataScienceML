```python
d = {'No':False, 'Yes':True}
df['International plan'] = df['International plan'].map(d)
df['International plan']
```

```python
df = df.replace({'Voice mail plan':d})
```

```python
df
```

```python
cols = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[cols].describe(percentiles = [])
```

```python
df.groupby(['Churn'])[cols].agg([np.mean, np.std, np.min, np.max])
```

```python
pd.crosstab(df['Churn'], df['International plan'])
```

```python
pd.crosstab(df['Churn'], df['Voice mail plan'], normalize=True)
```

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'],
              aggfunc='mean')
```

```python
df['total calls'] = df['Total day calls'] + df['Total eve calls'] + df['Total night calls'] + df['Total intl calls']
```

```python
df['total calls']
```

```python
df['total charge'] = df['Total day charge'] + df['Total eve charge'] + df['Total night calls'] + df['Total intl calls']
df['total charge']
```

```python
df
```

```python
df.drop(['total charge', 'total calls'], axis=1, inplace=True)
df.drop([1,2])
```

```python
pd.crosstab(df["Churn"], df['International plan'], margins=True)
```

```python
sns.countplot(x='International plan', hue='Churn', data=df)
```

```python
pd.crosstab(df['Churn'], df['Customer service calls'], margins=True)
```

```python
sns.countplot(x='Customer service calls', hue='Churn', data=df)
```

```python
df['many service calls'] = (df['Customer service calls'] > 3).astype('int')
pd.crosstab(df['Churn'], df['many service calls'], margins = True)
```

```python
sns.countplot(x='many service calls', hue= 'Churn', data=df)
```
